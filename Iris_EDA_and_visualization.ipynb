```python
X =df_iris.drop(columns=['target'] ) #inplace=True
```

```python
X
```

```python
y = df_iris['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,random_state=42) #test_size aswell
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier()
```

```python
model  = KNN.fit(X_train,y_train) # also just KNN.fit
```

```python
model.score(X_test,y_test)
```

```python
y_pred = KNN.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_test, y_pred, target_names=df.target_names)
print(report)
```
